**CREDIT CARD FRAUD DETECTION**

Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a pandas data frame
credit_card_data = pd.read_csv('/content/creditcard.csv.zip')
```

```python
# first 5 rows of the data set
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
# checking the number of missing values in each colum
credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction & fraudulent transaction
credit_card_data['Class'].value_counts()
```

This data is very unbalanced
If we  feed this to machine learning model bcz it can not detect fradulent case..



0-->Normal Transaction



1 -->Fradulent Transaction

```python
# seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical  measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transcations
credit_card_data.groupby('Class').mean()
```

Under sampling

Build a sample dataset containing similar distribution of normal transactions and fradulent Transactions

so we are going take 492 random legit transactions and 492 fraud transcations

Number of Fradulent transactions =  492

```python
legit_sample = legit.sample(n=492)
```

Concatenating two data frames

```python
new_dataset = pd.concat([legit_sample,fraud], axis =0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

splitting the data into Features and Targets

```python
X = new_dataset.drop(columns = 'Class', axis =1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

splitting the data into Training and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, stratify= Y,random_state =2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MODEL TRAINING

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with TRaining data
```

```python
model.fit(X_train,Y_train)
```

MODEL EVALUATION

ACCURACY SCORE

```python
# accuracy on Training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data : ',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on Testing data : ',test_data_accuracy)
```

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# Specifically import confusion_matrix and ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# ... (rest of your code) ...

# After the model evaluation part
cm = confusion_matrix(Y_test, X_test_prediction)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
# Sample Input Data for Testing the Model
sample_input = np.array([[150000.0, -1.2, 2.3, -1.4, 0.5, 1.2, -2.5, 0.3, -1.0, 1.1,
                          -0.5, 2.6, -1.2, -0.8, 0.7, -0.6, 1.0, -0.2, 0.4, -1.7,
                           0.5, -0.3, 1.2, 0.8, -1.1, 0.4, -0.7, 0.2, 0.5, 50.0]])

# Convert the input to match the model's format
sample_input_df = pd.DataFrame(sample_input, columns=X.columns)

# Make Prediction
prediction = model.predict(sample_input_df)

# Display Result
if prediction[0] == 0:
    print("Prediction: Legitimate Transaction")
else:
    print("Prediction: Fraudulent Transaction")
```

```python
from sklearn.metrics import precision_score

precision = precision_score(Y_test, X_test_prediction)
print("Precision:", precision)
```

```python
from sklearn.metrics import recall_score

recall = recall_score(Y_test, X_test_prediction)
print("Recall:", recall)
```

```python
# Assuming 'cm' is the confusion matrix calculated previously
specificity = cm[0, 0] / (cm[0, 0] + cm[0, 1])
print("Specificity:", specificity)
```
